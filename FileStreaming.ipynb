```python
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext
from pyspark import SparkContext
```

```python
sc = SparkContext(conf=SparkConf(), master='local[*]', appName='BookStream')
ssc = StreamingContext(sc, 2)
```

```python
texts = ssc.textFileStream(".\books\test")
```

```python
def aggregate_sum(new_values, total_sum):
    return sum(new_values) + (total_sum or 0)
```

```python
words = texts.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda x: (x, 1))
wordCounts = pairs.reduceByKey(aggregate_sum)
wordCounts.pprint()

ssc.start()

ssc.awaitTermination()
```
